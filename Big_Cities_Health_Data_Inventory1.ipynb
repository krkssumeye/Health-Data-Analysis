```python
health_data_female = health_data[(health_data["Gender"] == "Female")]  
#ilk öncelikle kadın hastaların, hastalık istatistikleri için cinsiyeti ayırıp yeni bir verisetine aktardık.
```

```python
health_data_female
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(25,14))    
sns.countplot(data=health_data_female, x="Race/ Ethnicity", hue="category",palette='Paired')
plt.title("Female Health Statistics")   
#kadın cinsiyetinin etnik kökenlerine göre hastalıkların istatistiklerini görselleştirdim.
```

```python
health_data_male = health_data[(health_data["Gender"] == "Male")]#Erkek hastaların, hastalık istatistikleri için cinsiyeti ayırıp yeni bir verisetine aktardık.
```

```python
plt.figure(figsize=(25,14))
sns.countplot(data=health_data_male, x="Race/ Ethnicity", hue="category",palette='Paired')
plt.title("Male Health Statistics")
#erkek cinsiyetinin etnik kökenlerine göre hastalıkların istatistiklerini görselleştirdim.
```

```python
health_data_both = health_data[(health_data["Gender"] == "Both")]
#her iki cinsiyeti temsil eden değer için de veriseti oluşturduk
```

```python
plt.figure(figsize=(25,14))
sns.countplot(data=health_data_both, x="Race/ Ethnicity", hue="category",palette='Paired')
plt.title("Both Health Statistics")
#etnik kökenlerine göre hastalıkların istatistiklerini görselleştirdim.
```

```python
health_data_hivaids = health_data[(health_data["category"] == "HIV/AIDS")]
#Daha çk detaylandırıp araştımamı kanser hastaları ve AIDS hastları üzerinde derinleştirmek istedim.
```

```python
plt.figure(figsize=(20,14))
sns.countplot(data=health_data_hivaids, x="Year", hue="Gender",palette='gist_stern')
plt.title("HIV/AIDS")
#öncelikle AIDS/HIV hastalığı için cinsiyetlerin yıllara göre artış değerini görselleştirdim.
```

```python
health_data_cancer = health_data[(health_data["category"] == "Cancer")]
```

```python
plt.figure(figsize=(20,14))
sns.countplot(data=health_data_cancer, x="Year", hue="Gender",palette='gist_stern')
plt.title("Cancer"),
#Cancer hastalığı için cinsiyetlerin yıllara göre artış değerini görselleştirdim.
```

```python
plt.figure(figsize=(60,14))
sns.countplot(data=health_data_cancer, x="Place", hue="Gender",palette='gist_heat_r')
plt.title("Cancer")
#Daha sonra Kanser hastalığının yerleşkelere göre cinsiyet istatiğini görselleştirdim.
```

```python
plt.figure(figsize=(60,14))
sns.countplot(data=health_data_hivaids, x="Place", hue="Gender",palette='gist_heat_r')
plt.title("HIV/AIDS")
#HIV/AIDS hastalığının yerleşkelere göre cinsiyet istatiğini görselleştirdim.
```
